```python
tradingset =Tesla_info[Tesla_info["日期"]>TradingDate] 

stackdf=signal.stack()
# stackdf.columns.name="value"
# print(stackdf)
# print(stackdf.columns)
# print(stackdf.reset_index("胜负"))
# sf_signal= stackdf.reset_index("胜负")
sf_signal= stackdf.reset_index()
print(sf_signal)

target =pd.merge(tradingset.round(1) ,sf_signal.round(1) ,how='inner',on ='收盘')
print(target.sort_values('收盘') )
# print(type(pd.merge(tradingset,sf_signal,how='inner',on ='收盘')))


# stack_redf=stackdf.unstack(0)
# print(stack_redf)
# print(stack_redf.stack())
# print(stackdf.index)
```

```python
fig, ax = plt.subplots()
def plot_func(group):
    global ax
    color = 'r' if (group["胜负"] == "Win").all() else 'g'
    lw = 2.0
    ax.plot(group["日期"], group["收盘"], c=color,linestyle='dashed',marker='o', markerfacecolor=color, markersize=12, linewidth=lw)
#Tesla_info.groupby("胜负").apply(plot_func)

startdate = "2019-04-17 "
enddate ="2019-08-30"
Tesla_info[(Tesla_info["日期"]>=startdate)&(Tesla_info["日期"]<=enddate)].groupby("胜负").apply(plot_func)
#Tesla_info.groupby("胜负").apply(plot_func)

plt.xticks(rotation=45)
```
